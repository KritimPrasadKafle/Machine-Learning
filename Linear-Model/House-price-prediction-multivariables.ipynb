```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
data = {
    "areas": [2000,3000,3200,3600,4000],
    "bedrooms":[3.0,4.0, " ", 3.0, 5.0],
    "age":[20,15,18,30,8],
    "price":[550000,565000,610000,595000,760000],
}
```

```python
df = pd.DataFrame(data)
```

```python
df["bedrooms"]= df["bedrooms"].replace(" ", float("nan"))
```

```python
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
```

```python
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
```

```python
df.bedrooms
```

```python
df
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['areas','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
reg.predict([[4000,6,50]])
```
