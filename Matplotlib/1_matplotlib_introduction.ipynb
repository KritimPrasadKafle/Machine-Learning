```python
import numpy as np
x = np.arange(12).reshape(3,4)
y = np.arange(3,15).reshape(3,4)
x
y
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='green', linewidth=5, linestyle = 'dotted')
```
