```python
import pandas as pd

data = {
    'day': ["1/1/2017", "1/2/2017", "1/4/2017", "1/5/2017", "1/6/2017", "1/7/2017"],
    'temperature': [32, 35, 28, "", 32, ""],
    'windspeed': [6, 7, 2, '', 23, ""],
    'event': ["Rain", "Sunny", "Snow", "", "Cloudy", ""],
}

# Create the DataFrame
df = pd.DataFrame(data)

# Replace empty strings with NaN
df = df.replace("", float("nan"))

print(df)
```

```python
df.fillna(0)
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event': "no event"
})
```

```python
new_df
```

```python
new_df = df.fillna(method = "bfill")
new_df
```

```python
new_df = df.fillna(method = "ffill")
new_df
```

```python
new_df = df.fillna(method = "ffill", limit = 1)
new_df
```

```python
new_df = df.interpolate()
```

```python
new_df
```

```python
new_df = df.dropna()
```

```python
new_df
```

```python
new_df = df.dropna(how = "all")
```
